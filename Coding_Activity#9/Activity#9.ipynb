```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Try to implement the linear transformations using spancs using the scatter plot view. Just do at least one example of linear transformation for a 2D space.

```python
X = np.array([
    [2,0],
    [0,2]
])

c = np.arange(0,1,0.15)

plt.scatter(c*X[0,0],c*X[1,0], color='red')
plt.scatter(c*X[0,1],c*X[1,1], color='blue')


plt.xlim(-2,4)
plt.ylim(-2,4)
plt.grid()
plt.show()
```

```python
Sc = np.array([
    [0.7,0],
    [0,0.7]
])

c = np.arange(0,1,0.30)

plt.scatter(c*Sc[0,0],c*Sc[1,0], color='red')
plt.scatter(c*Sc[0,1],c*Sc[1,1], color='blue')


plt.xlim(-2,4)
plt.ylim(-2,4)
plt.grid()
plt.show()
```

```python
S = np.array([
    [2,0],
    [1,2]
])

c = np.arange(0,1,0.15)

plt.scatter(c*S[0,0],c*S[1,0], color='red')
plt.scatter(c*S[0,1],c*S[1,1], color='blue')


plt.xlim(-2,4)
plt.ylim(-2,4)
plt.grid()
plt.show()
```

```python
S1 = np.array([
    [-2,0],
    [0,-2]
])

c = np.arange(0,1,0.15)

plt.scatter(c*S1[0,0],c*S1[1,0], color='red')
plt.scatter(c*S1[0,1],c*S1[1,1], color='blue')


plt.xlim(-2,4)
plt.ylim(-2,4)
plt.grid()
plt.show()
```
