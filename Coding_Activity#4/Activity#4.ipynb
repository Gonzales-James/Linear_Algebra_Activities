```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<b><h3>Task 1<h3><b>

```python
def explicit_func(X):
    X = np.sqrt((X[0]**2)+(X[1]**2)+(X[2]**2)+(X[3]**2)+(X[4]**2))
    print(X)
A = np.array([1,2,3,4,5])
B = np.array([2,2,1,9,3])
C = np.array([6,2,7,4,6])
D = np.array([5,3,5,7,8])
E = np.array([6,4,3,6,4])
F = np.array([2,4,6,8,8])

print("Using the explicit function: ")
explicit_func(A)
explicit_func(B)
explicit_func(C)
explicit_func(D)
explicit_func(E)
explicit_func(F)
print("\n")
print("Using the NumPy's preset function: ")
print(np.linalg.norm(A))
print(np.linalg.norm(B))
print(np.linalg.norm(C))
print(np.linalg.norm(D))
print(np.linalg.norm(E))
print(np.linalg.norm(F))
```

<b><h3>Task 2<h3><b>

```python
def exp_func(A, B):
    Product = ((A[0]*B[0]) + (A[1]*B[1]) + (A[2]*B[2]) + (A[3]*B[3]) + (A[4]*B[4]))
    print(Product)
A = np.array([1,2,3,4,5])
B = np.array([2,2,1,9,3])

C = np.array([6,2,7,4,6])
D = np.array([5,3,5,7,8])

E = np.array([6,4,3,6,4])
F = np.array([2,4,6,8,8])

G = np.array([2,4,6,3,1])
H = np.array([6,4,2,5,6])

I = np.array([8,5,3,7,1])
J = np.array([9,0,6,3,0])

print("Using the explicit function: ")
exp_func(A, B)
exp_func(C, D)
exp_func(E, F)
exp_func(G, H)
exp_func(I, J)
print("\n")
print("Using the NumPy's preset function: ")
print(np.inner(A, B))
print(np.inner(C, D))
print(np.inner(E, F))
print(np.inner(G, H))
print(np.inner(I, J))
```

<b><h3>Task 3<h3><b>

```python
def formula(A,B,C):
    Total = ((A@A + B@B + C@C) * (A * (B + A * B) / C)) * np.linalg.norm(A+B+C)
    print(Total)
    
A = np.array([-0.4, 0.3, -0.6])
B = np.array([-0.2, 0.2, 1])
C = np.array([0.2, 0.1, -0.5])

print("My Output: ")
formula(A,B,C)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim([-1, 2])
ax.set_ylim([-1, 2])
ax.set_zlim([-1, 2])

ax.quiver(0, 0, 0, 0.34769805, 1.13001866, 0.6953961, colors='red')

plt.show()


print("\nExpected Output: ")
print(np.array([0.34769805, 1.13001866, 0.6953961 ]))
```
