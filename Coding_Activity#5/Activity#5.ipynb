```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<b><h3>Task 1<h3><b><br>
$$ A = 5x + 2y $$

```python
A = np.array([5,2])
print("Rank: ",A.ndim)
plt.xlim(0,10)
plt.ylim(0,10)

plt.quiver(0,0, A[0], A[1], angles='xy', scale_units='xy',scale=1, color='red') 
plt.show()
```

<b><h3>Task 2<h3><b><br>
$$ B = \left\{\begin{array}\\ 8x + 2y + 3z \\ x + 6y + 4z\end{array}\right. $$

```python
B = np.array([[8,2,3],
              [1,6,4]])
print("Rank: ",B.ndim)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])

ax.quiver(0, 0, 0, B[0,0], B[0,1], B[0,2], arrow_length_ratio=0.05, colors='red')
ax.quiver(0, 0, 0, B[1,0], B[1,1], B[1,2], arrow_length_ratio=0.05, colors='blue')
plt.show()
```
